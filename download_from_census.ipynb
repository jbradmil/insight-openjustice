```python
# set the parameters we want
years=[2015]
## You'll need a Census API key 
## Set an alias to the key in (e.g.) your .bashrc file
## export CENSUS_API_KEY=####KEY####
api_key = os.environ.get('CENSUS_API_KEY')
```

This will create the directories in which we'll store the downloaded tables, then pull the lists of variables for each table and create a dictionary for them. You shouldn't need to modify this cell, but if you want to change the selection of tables, you'll need to edit the build_variable_dict function in utils/acs_tools.py. Currently, it only looks for tables whose names begin with 'C' (if (table[0]!='C'...). These are fairly high-level summary tables. If you want more finely segmented data, you may wish to switch to the 'B' tables.

```python
base_dir=build_output_directories(years)
variable_lists=pull_variable_lists(years)
var_dict=build_variable_dict(variable_lists)
```

Here's where we download and save the data from the selected tables. This may take a while to run. If you want to test it on just one table, go into utils/acs+tools.py, then uncomment the break statement.

```python
download_and_save_data(var_dict, api_key)
```
